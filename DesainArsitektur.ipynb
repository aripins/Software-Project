```python
from graphviz import Digraph
from IPython.display import Image, display

# Initialize Digraph
arch_diagram = Digraph('System_Architecture', format='png')
arch_diagram.attr(rankdir='TB', size='20,20', bgcolor='white')  # White background for the entire diagram

# Define styles for clusters and nodes with color and transparency
cluster_style = {
    'style': 'filled',
    'fontname': 'Helvetica',
    'fontsize': '12',
    'fillcolor': '#e0f7fa',  # Light cyan color for the background of layers
    'color': 'black',  # Border color of the clusters
}

node_style = {
    'shape': 'box',
    'fontname': 'Helvetica',
    'fontsize': '10',
    'style': 'filled',
    'fillcolor': '#ffffff',  # White background for nodes
    'color': '#4CAF50',  # Green border for nodes
    'fontcolor': 'black',  # Text color inside nodes
}

# Presentation Layer
with arch_diagram.subgraph(name='cluster_presentation') as presentation:
    presentation.attr(label='PRESENTATION LAYER', **cluster_style)
    presentation.node('Web Browser', 'Web Browser', **node_style)

# Service Layer
with arch_diagram.subgraph(name='cluster_service') as service:
    service.attr(label='SERVICE LAYER', **cluster_style)
    service.node('SkillLink Backend', 'SkillLink Backend (HTML5 & Bootstrap API)', **node_style)
    service.node('Authentication Service', 'Authentication Service', **node_style)
    service.node('Recommendation Engine', 'Recommendation Engine', **node_style)
    service.node('Notification Service', 'Notification Service', **node_style)
    service.node('Analytics Service', 'Analytics Service', **node_style)
    service.node('File Management', 'File Management Service', **node_style)

# Data Layer
with arch_diagram.subgraph(name='cluster_data') as data:
    data.attr(label='DATA LAYER', **cluster_style)
    data.node('PostgreSQL Database', 'PostgreSQL Database', **node_style)
    data.node('Redis Cache', 'Redis Cache', **node_style)
    data.node('File Storage', 'File Storage (AWS S3)', **node_style)
    data.node('Data Warehouse', 'Data Warehouse (BigQuery)', **node_style)

# External Systems
with arch_diagram.subgraph(name='cluster_external') as external:
    external.attr(label='EXTERNAL SYSTEMS', **cluster_style)
    external.node('LinkedIn API', 'LinkedIn API', **node_style)
    external.node('GitHub Repository', 'GitHub Repository', **node_style)

# Define connections
arch_diagram.edge('Web Browser', 'SkillLink Backend', label='HTTPS')
arch_diagram.edge('SkillLink Backend', 'Authentication Service', label='Auth Token')
arch_diagram.edge('SkillLink Backend', 'Recommendation Engine', label='Skill Matching')
arch_diagram.edge('SkillLink Backend', 'Notification Service', label='Push Notifications')
arch_diagram.edge('SkillLink Backend', 'Analytics Service', label='User Metrics')
arch_diagram.edge('SkillLink Backend', 'File Management', label='File Operations')
arch_diagram.edge('SkillLink Backend', 'PostgreSQL Database', label='SQL Queries')
arch_diagram.edge('SkillLink Backend', 'Redis Cache', label='Caching')
arch_diagram.edge('SkillLink Backend', 'File Storage', label='Store Files')
arch_diagram.edge('Analytics Service', 'Data Warehouse', label='Data Insights')
arch_diagram.edge('Notification Service', 'LinkedIn API', label='LinkedIn Notifications')
arch_diagram.edge('Notification Service', 'GitHub Repository', label='GitHub Sync')

# Save and render the diagram
output_file = 'SkillLink_System_Architecture_With_Color'
arch_diagram.render(output_file, cleanup=True)

# Display the diagram in Colab
display(Image(f'{output_file}.png'))

print("Colorful system architecture diagram for SkillLink created successfully and displayed.")
```

```python
from graphviz import Digraph
from IPython.display import Image, display

# Initialize Diagram
diagram = Digraph('SkillLink_High_Level', format='png')
diagram.attr(rankdir='LR', size='12,12', fontname='Helvetica')

# Diagram Background Style
diagram.attr(bgcolor='#F9F9F9', style='filled', fillcolor='#EAF2F8')

# Title
diagram.attr(label='SkillLink Network System (High-Level Design)\n', labelloc='t', fontsize='20', fontcolor='#1A5276')

# Style Definitions
cluster_style = {
    'style': 'rounded, filled',
    'color': '#2E86C1',
    'fontname': 'Helvetica-Bold',
    'fontsize': '16',
    'fillcolor': '#D4E6F1'
}

node_style = {
    'shape': 'plaintext',
    'fontname': 'Helvetica',
    'fontsize': '14',
    'fontcolor': '#154360'
}

arrow_style = {
    'color': '#2874A6',
    'penwidth': '2'
}

# Function to create HTML-like tables for nodes
def create_table(title, points, bgcolor='#D4E6F1'):
    header = f'<tr><td align="center" bgcolor="#5DADE2"><b>{title}</b></td></tr>'
    rows = ''.join([f'<tr><td align="left" bgcolor="{bgcolor}">• {point}</td></tr>' for point in points])
    return f'<<table border="0" cellborder="0" cellspacing="0">{header}{rows}</table>>'

# Define Nodes with Uniform Blue Theme
diagram.node('UI1', create_table('User Interface', ['Browse Projects', 'Manage Profile', 'Track Recommendations']), **node_style)
diagram.node('Auth1', create_table('Authentication Service', ['Sign Up', 'Log In', 'Reset Password']), **node_style)
diagram.node('Project1', create_table('Project Management', ['Create Project', 'Assign Roles', 'Track Progress']), **node_style)
diagram.node('Recommendation1', create_table('Recommendation Engine', ['Analyze Skills', 'Suggest Projects', 'Update Recommendations']), **node_style)
diagram.node('Notification1', create_table('Notification Service', ['Send Updates', 'Alert Notifications']), **node_style)
diagram.node('Database1', create_table('Database', ['Store User Data', 'Store Project Data', 'Store Logs', 'Store Notifications']), **node_style)
diagram.node('External1', create_table('External Systems', ['GitHub Integration', 'LinkedIn API']), **node_style)

# Define Edges and Connections with Cleaner Layout
diagram.edge('UI1', 'Auth1', label='Authenticate', **arrow_style)
diagram.edge('UI1', 'Recommendation1', label='Fetch Suggestions', **arrow_style)
diagram.edge('UI1', 'Project1', label='Create/Manage Projects', **arrow_style)
diagram.edge('Recommendation1', 'Database1', label='Store Logs', **arrow_style)
diagram.edge('Project1', 'Database1', label='Save Projects', **arrow_style)
diagram.edge('Notification1', 'Database1', label='Save Notifications', **arrow_style)
diagram.edge('External1', 'Database1', label='Sync Data', **arrow_style)
diagram.edge('Notification1', 'External1', label='Send Alerts', **arrow_style)

# Render diagram and display
diagram.render('SkillLink_Network_Design_Blue', cleanup=True)
display(Image(filename='SkillLink_Network_Design_Blue.png'))
```
