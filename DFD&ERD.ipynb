```python
from graphviz import Digraph
from IPython.display import Image, display
import os

# Initialize the DFD graph
dfd = Digraph('DFD_Level1', filename='dfd_skilllink_network', format='png')
dfd.attr(rankdir='LR', size='10', color='blue')

# Set default colors for all elements
dfd.attr('node', color='blue', fontcolor='black')  # Nodes: blue outline, black text
dfd.attr('edge', color='blue', fontcolor='black')  # Edges: blue line, black label

# External Entities
dfd.node('BackOffice', 'Back Office', shape='box', color='blue', fontcolor='black')
dfd.node('Users', 'Users (PM & Collaborators)', shape='box', color='blue', fontcolor='black')

# Processes
dfd.node('P1', '1. Create Project', shape='circle', color='blue', fontcolor='black')
dfd.node('P2', '2. Manage Team', shape='circle', color='blue', fontcolor='black')
dfd.node('P3', '3. Manage Tasks', shape='circle', color='blue', fontcolor='black')
dfd.node('P4', '4. Monitor Progress', shape='circle', color='blue', fontcolor='black')

# Data Stores
dfd.node('D1', 'Project Data', shape='rect', color='blue', fontcolor='black')
dfd.node('D2', 'Team Data', shape='rect', color='blue', fontcolor='black')
dfd.node('D3', 'Task Data', shape='rect', color='blue', fontcolor='black')
dfd.node('D4', 'Project Report', shape='rect', color='blue', fontcolor='black')

# Data Flows (Edges)
# From Back Office
dfd.edge('BackOffice', 'P1', label='Project Data', color='blue', fontcolor='black')
dfd.edge('P4', 'BackOffice', label='Project Report', color='blue', fontcolor='black')

# Process 1: Create Project
dfd.edge('P1', 'D1', label='Save Project Data', color='blue', fontcolor='black')
dfd.edge('P1', 'Users', label='Project Information', color='blue', fontcolor='black')

# Process 2: Manage Team
dfd.edge('Users', 'P2', label='Team Request', color='blue', fontcolor='black')
dfd.edge('P2', 'D2', label='Save Team Data', color='blue', fontcolor='black')
dfd.edge('P2', 'Users', label='Team Information', color='blue', fontcolor='black')

# Process 3: Manage Tasks
dfd.edge('Users', 'P3', label='Task Input', color='blue', fontcolor='black')
dfd.edge('P3', 'D3', label='Save Task Data', color='blue', fontcolor='black')
dfd.edge('P3', 'Users', label='Task Status', color='blue', fontcolor='black')

# Process 4: Monitor Progress
dfd.edge('D3', 'P4', label='Task Data', color='blue', fontcolor='black')
dfd.edge('P4', 'D4', label='Generate Project Report', color='blue', fontcolor='black')

# Render and display the graph
output_path = dfd.render(filename='/content/dfd_skilllink_network_hd', format='png', cleanup=True)
print("The file is saved at:", output_path)

# Check if file exists
print("Does the file exist?:", os.path.exists(output_path))

# Display the high-quality image
display(Image(output_path))
```

```python
# Install Graphviz jika belum ada
!apt-get install graphviz
!pip install graphviz

# Import library
from graphviz import Digraph
from IPython.display import Image, display

# Create ERD for SkillLink with horizontal layout
erd = Digraph(comment='ERD SkillLink', format='png')
erd.attr(rankdir='LR')  # Set layout to Left-to-Right

# Table: Users
erd.node('Users', '''<
<TABLE BORDER="1" CELLBORDER="0" CELLSPACING="0" CELLPADDING="5">
  <TR><TD BGCOLOR="#cce5ff"><B>Users</B></TD></TR>
  <TR><TD ALIGN="LEFT">IdUser (PK)</TD></TR>
  <TR><TD ALIGN="LEFT">Name</TD></TR>
  <TR><TD ALIGN="LEFT">Role</TD></TR>
  <TR><TD ALIGN="LEFT">Email</TD></TR>
  <TR><TD ALIGN="LEFT">Password</TD></TR>
</TABLE>>''', shape='plaintext')

# Table: Projects
erd.node('Projects', '''<
<TABLE BORDER="1" CELLBORDER="0" CELLSPACING="0" CELLPADDING="5">
  <TR><TD BGCOLOR="#cce5ff"><B>Projects</B></TD></TR>
  <TR><TD ALIGN="LEFT">IdProject (PK)</TD></TR>
  <TR><TD ALIGN="LEFT">ProjectName</TD></TR>
  <TR><TD ALIGN="LEFT">Description</TD></TR>
  <TR><TD ALIGN="LEFT">StartDate</TD></TR>
  <TR><TD ALIGN="LEFT">EndDate</TD></TR>
  <TR><TD ALIGN="LEFT">IdCreator (FK)</TD></TR>
</TABLE>>''', shape='plaintext')

# Table: Tasks
erd.node('Tasks', '''<
<TABLE BORDER="1" CELLBORDER="0" CELLSPACING="0" CELLPADDING="5">
  <TR><TD BGCOLOR="#cce5ff"><B>Tasks</B></TD></TR>
  <TR><TD ALIGN="LEFT">IdTask (PK)</TD></TR>
  <TR><TD ALIGN="LEFT">TaskName</TD></TR>
  <TR><TD ALIGN="LEFT">Details</TD></TR>
  <TR><TD ALIGN="LEFT">Deadline</TD></TR>
  <TR><TD ALIGN="LEFT">TaskStatus</TD></TR>
  <TR><TD ALIGN="LEFT">IdProject (FK)</TD></TR>
  <TR><TD ALIGN="LEFT">IdCollaborator (FK)</TD></TR>
</TABLE>>''', shape='plaintext')

# Relationships
erd.edge('Users', 'Projects', label='Creates Projects', color='black', penwidth='2.0')
erd.edge('Projects', 'Tasks', label='Contains Tasks', color='black', penwidth='2.0')
erd.edge('Users', 'Tasks', label='Works on Tasks', color='black', penwidth='2.0')

# Render and save the ERD diagram
erd.render('/content/ERD_SkillLink_Horizontal', cleanup=True)

# Display the image directly in Colab
display(Image('/content/ERD_SkillLink_Horizontal.png'))

print("ERD for SkillLink with horizontal table layout has been successfully generated and displayed.")
```
